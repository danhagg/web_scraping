# Use python (Beautiful Soup) to extract all html

````python
from bs4 import BeautifulSoup
import requests

# create a beautiful soup object from file
# soup object of parsed html
with open('Attorneys - Lewis Brisbois Bisgaard & Smith LLP.html') as html_file:
    soup = BeautifulSoup(html_file, 'lxml')

# can print out normal or pretty(indents)
print(soup.prettify())
````

# a-tag gets me name, position, location

````python
# This will get the div
div_container = soup.find('ul', class_="attorneys list")  

alist = []
# Then search in that div_container for all a tags
for atag in div_container.find_all('a'):
    # strip all whitespace
    descriptions = list(atag.stripped_strings)
    alist.append(descriptions)

# Take the 1th item and every 3th onwards
main_info = alist[1::3]

milist = []
for i in main_info:
    #milist.append([i[0], i[2]])
    milist.append(i[0::2])
    
print(milist)
````

# phone

````python
# This will get the div
div_container = soup.find('ul', class_="attorneys list")  

phlist = []

for phtag in div_container.find_all('div', class_="phone"):
    descriptions = list(phtag.stripped_strings)
    phlist.append(descriptions)

pplist = []
for i in phlist:
    pplist.append(i[0])
    
print(pplist)
````

````python
div_container = soup.find('ul', class_="attorneys list")  

elist = []
for etag in div_container.find_all('div', class_="email"):
    descriptions = list(etag.stripped_strings)
    elist.append(descriptions)

eelist = []
for i in elist:
    eelist.append(i[0])
    
print(eelist)
````

# Export lists to pandas

````python
import pandas as pd
attorney_list = pd.DataFrame(
    {'main_info': milist,
     'lphone': pplist,
     'email': eelist
    })
````
